```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['txn_month'] = df['transaction_date'].dt.month
display(df.head())
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['txn_quarter'] = df['transaction_date'].dt.quarter
display(df)
```

```python
df['txn_dateofweek'] = df['transaction_date'].dt.dayofweek
display(df)
```

```python
df['revenue_per_unit'] = df['price_per_unit'] * df['quantity']
display(df)
```

```python
df['is_bulk'] = df['quantity'] >= 7
display(df)
```

```python
df['spent_segment'] = pd.qcut(df['revenue_per_unit'], q=3, labels=['Low', 'Medium', 'High'])
display(df)
```

```python
df['discount_flag'] = df['discount_applied'].astype(int)
display(df)
```
